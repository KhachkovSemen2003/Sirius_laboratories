Равносторонний (все стороны равны)

Разносторонний (все стороны разные)

```python
def tri(mark1:int, mark2:int, mark3:int) -> str:
    if ((mark1 + mark2) <= mark3) or ((mark1 + mark3) <= mark2) or ((mark3 + mark2) <= mark1):
        return "no triangle"
    elif mark1 == mark2 == mark3:
        return "равносторонний"
    elif mark1 == mark2 or mark1 == mark3 or mark3 == mark2:
        return "равнобедренный"
    else:
        return "классический"
print(tri(1,2,1))
```

```python
def simple(n):
    smpl = []
    for i in range(2, n + 1):
        flag = 0
        for j in range(2, i - 1):
            if i % j == 0:
                flag = 1
                break
        if flag == 0:
            smpl.append(i)
    return smpl
print(simple(23))
```

Задание 3.3: Кастомные исключения
Создайте систему кастомных исключений для банковского приложения:

InsufficientFundsError (недостаточно средств)
InvalidAccountError (неверный номер счета)
TransactionLimitError (превышен лимит транзакций)
Реализуйте класс BankAccount с методами deposit, withdraw, transfer
Используйте ваши кастомные исключения

```python
class InsufficientFundsError(Exception):
    def __init__(self, message="Недостаточно средств на счете"):
        self.message = message
        super().__init__(self.message)
        
class InvalidAccountError(Exception):
    def __init__(self, acc_number, message="Неверный номер счета"):
        self.acc_number = acc_number
        self.message = message
        super().__init__(self.message)
        
class TransactionLimitError(Exception):
    def __init__(self, message="превышен лимит транзакций"):
        self.message = message
        super().__init__(self.message)

class BankAccount:
    def __init__(self, acc_number, balance = 0, lim_tranz = 100):
        if not self._val_acc_num(acc_number):
            raise InvalidAccountError(acc_number)
        self.acc_number = acc_number
        self.balance = balance
        self.lim_tranz = lim_tranz
    def _val_acc_num(self, acc_number):
        if not isinstance(acc_number, int):
            return False
        if len(str(acc_number)) != 15:
            return False
        return True
    def deposit(self, dep:int) -> int:
        if dep < 0:
            raise ValueError("депозит должен быть больше нуля")
        self.balance += dep
        self.lim_tranz -=1
        return self.balance
    def withdraw(self, withdr:int) -> int:
        if withdr < 0:
            raise ValueError("сумма вывода должна быть больше нуля")
        if withdr > self.balance:
            raise InsufficientFundsError("баланс меньше суммы вывода")
        if self.lim_tranz <= 0:
            raise TransactionLimitError("превышен лимит транзакций")
        self.balance -= withdr
        self.lim_tranz -= 1
        return self.balance
    def transfer(self, target_account:BankAccount, delive:int) -> int:
        if delive < 0:
            raise ValueError("сумма отправки должна быть больше нуля")
        if delive > self.balance:
            raise InsufficientFundsError("сумма отправки превышает баланс")
        if self.lim_tranz <= 0:
            raise TransactionLimitError("превышен лимит транзакций")
        self.balance -= delive
        target_account.balance += delive
        self.lim_tranz -= 1
        target_account.lim_tranz -= 1
        return self.balance
```

Задание 4.1: Кастомный итератор
Создайте класс, который реализует итератор для последовательности Фибоначчи:

Реализуйте методы iter и next
Ограничьте количество итераций параметром max_iterations
Обработайте исключение StopIteration

```python
class Iterator_Fibonachi:
    def __init__(self, max_iterations):
        if max_iterations <= 0:
            raise ValueError("количество итераций должно быть больше нуля")
        self.max_iterations = max_iterations
        self.act_iterations = 0
        self.a = 0
        self.b = 1
    def __iter__(self):
        return self
    def __next__(self) -> int:
        result = None
        if self.act_iterations >= self.max_iterations:
            raise StopIteration("превышено максимальное количество итераций")
        if self.act_iterations == 0:
            result = self.a
        elif self.act_iterations == 1:
            result = self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            result = self.b
        self.act_iterations += 1
        return result
```

Задание 4.2: Генератор простых чисел¶
Напишите генератор, который выдает простые числа:

Используйте yield для генерации чисел
Реализуйте эффективный алгоритм проверки простоты
Добавьте возможность ограничить количество генерируемых чисел

```python
from typing import Iterator

def prime_numbers_generator(limit: int) -> Iterator[int]:
    
    def simple(n: int) -> int:
        if n < 2:
            return 0
        if n == 2:
            return 1
        if n % 2 == 0:
            return 0
        
        max_divisor = int(n ** 0.5) + 1
        for i in range(3, max_divisor, 2):
            if n % i == 0:
                return 0
        return 1
    
    count = 0
    n = 2
    
    while count < limit:
        if simple(n):
            yield n
            count += 1
        n += 1
```

Задание 4.3: Генератор с send()
Создайте генератор, который:

Генерирует последовательность квадратов чисел
Принимает через send() новое начальное значение
Может быть сброшен к новому начальному значению
Логирует все операции

```python
from typing import Generator
import logging

logging.basicConfig(level=logging.INFO, force=True)

def kvadraty_generator(start:int = 0) -> Generator[int, int | None, None]:
    logging.info("Функция запущена")
    current = start
    try:
        while True:
            square = current ** 2
            logging.info(f"Вычислен квадрат: {current}² = {square}")
            new_start = yield square
            logging.info(f"Получено значение через send(): {new_start}")
            if new_start is not None:
                current = new_start
                logging.info(f"Сброс генератора к новому началу: {current}")
            else:
                current += 1
    except GeneratorExit:
        logging.info("Генератор завершил работу")
        raise
gen = kvadraty_generator(2)
print(next(gen))
gen.close()
```

Задание 4.4: Рекурсивный генератор
Реализуйте генератор для обхода дерева:

Создайте класс TreeNode с дочерними узлами
Напишите рекурсивный генератор для обхода дерева в глубину
Добавьте возможность обхода в ширину

```python
from typing import Any, Dict, Optional,  Union, Iterator
from collections import deque

class TreeNode:
    
    def __init__(self, structure: Dict[str, Any] = None):
        if structure is None:
            structure = {'data': 0, 'left': None, 'right': None}
        
        if 'left' not in structure:
            structure['left'] = None
        if 'right' not in structure:
            structure['right'] = None
            
        if isinstance(structure.get('left'), dict):
            structure['left'] = TreeNode(structure['left'])
        if isinstance(structure.get('right'), dict):
            structure['right'] = TreeNode(structure['right'])
            
        self.structure = structure
    @property
    def data(self) -> Any:
        return self.structure['data']
    
    @data.setter
    def data(self, value: Any) -> None:
        self.structure['data'] = value
    
    @property
    def left(self) -> Optional['TreeNode']:
        left_val = self.structure['left']
        if isinstance(left_val, dict):
            self.structure['left'] = TreeNode(left_val)
        return self.structure['left']
    
    @left.setter
    def left(self, node: Optional[Union['TreeNode', Dict[str, Any]]]) -> None:
        if isinstance(node, dict):
            node = TreeNode(node)
        self.structure['left'] = node
    
    @property
    def right(self) -> Optional['TreeNode']:
        right_val = self.structure['right']
        if isinstance(right_val, dict):
            self.structure['right'] = TreeNode(right_val)
        return self.structure['right']
    
    @right.setter
    def right(self, node: Optional[Union['TreeNode', Dict[str, Any]]]) -> None:
        if isinstance(node, dict):
            node = TreeNode(node)
        self.structure['right'] = node

def depth_gen(node: Optional[TreeNode]) -> Iterator[TreeNode]:
    if node is None:
        return
    yield node
    yield from depth_gen(node.left)
    yield from depth_gen(node.right)
    
def width_gen(node: Optional[TreeNode]) -> Iterator[TreeNode]:
    if node is None:
        return
    queue = deque([node])
    while queue:
        current = queue.popleft()
        yield current
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
            
root = TreeNode({
    'data': 1, 
    'left': {
        'data': 2, 
        'left': {'data': 4}, 
        'right': {'data': 5}
    }, 
    'right': {'data': 3}
})

print( [node.data for node in depth_gen(root)])
print([node.data for node in width_gen(root)])
```

Задание 5.2: Логирование с контекстом
Реализуйте систему логирования с контекстной информацией:

Добавляйте к каждому логу информацию о пользователе
Включайте timestamp и уникальный ID запроса
Используйте структурированное логирование (JSON формат)
Создайте декоратор для автоматического логирования функций

```python
import logging
import json
from datetime import datetime
from functools import wraps
import uuid
from typing import Any, Dict, Callable
import unittest

class StructuredLogger:
    def __init__(self):
        self.logger = logging.getLogger('structured')
        self.logger.setLevel(logging.DEBUG)

        handler = logging.StreamHandler()
        formatter = logging.Formatter('%(message)s')
        handler.setFormatter(formatter)
        self.logger.addHandler(handler)
    
    def log(self, level: str, message: str, **context: Any) -> None:

        log_data = {
            "timestamp": datetime.now().isoformat(),
            "level": level.upper(),
            "message": message,
            "request_id": str(uuid.uuid4()),
            "context": context
        }
        
        log_json = json.dumps(log_data, ensure_ascii=False)
        
        if level == 'debug':
            self.logger.debug(log_json)
        elif level == 'info':
            self.logger.info(log_json)
        elif level == 'error':
            self.logger.error(log_json)


structured_logger = StructuredLogger()

def log_function(logger: StructuredLogger):
    def decorator(func: Callable):
        def wrapper(*args, **kwargs):
            user_info = {}
            for arg in args:
                if isinstance(arg, dict) and 'user' in arg:
                    user_info = arg.get('user', {})
                    break
            
            for key, value in kwargs.items():
                if 'user' in key.lower():
                    user_info = value if isinstance(value, dict) else {'username': value}
                    break
            
            logger.log('info', 
                      f"Начало выполнения функции {func.__name__}",
                      function=func.__name__,
                      user=user_info,
                      args=str(args),
                      kwargs=str(kwargs))
            
            try:
                result = func(*args, **kwargs)
                logger.log('info',
                          f"Функция {func.__name__} выполнена успешно",
                          function=func.__name__,
                          user=user_info,
                          result=str(result) if result is not None else "None")
                return result
            except Exception as e:
                logger.log('error',
                          f"Ошибка в функции {func.__name__}: {str(e)}",
                          function=func.__name__,
                          user=user_info,
                          error_type=type(e).__name__)
                raise
        
        return wrapper
    return decorator

@log_function(structured_logger)
def example_processing(user_data: Dict[str, Any], item: str) -> str:
    """Пример функции для тестирования"""
    structured_logger.log('info', 'Обработка элемента', user=user_data, item=item)
    return f"Обработано: {item}"


print("1. Тест ручного логирования:")
structured_logger.log('info', 'Ручной тест', manual_test=True)
print("Ручное логирование работает")
print()
    
print("2. Тест декоратора:")
user = {'name': 'test_user', 'id': 123}
result = example_processing(user, 'тестовый_элемент')
print(f"Результат: {result}")
```

```python
import os
import logging
import platform
import time
import sys
from typing import Callable, Any
import functools

# Пороги для предупреждений
TIME_THRESHOLD = 0.1  # секунды
BLOCKS_THRESHOLD = 1000  # блоков памяти

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('performance.log'),
        logging.StreamHandler()
    ]
)

logger = logging.getLogger(__name__)

def monitor_performance(func: Callable) -> Callable:
    """Декоратор для мониторинга производительности функций."""
    
    @functools.wraps(func)
    def wrapper(*args, **kwargs) -> Any:
        start_blocks = sys.getallocatedblocks()
        
        start_time = time.time()
        
        try:
            result = func(*args, **kwargs)
            
            execution_time = time.time() - start_time
            
            end_blocks = sys.getallocatedblocks()
            blocks_used = end_blocks - start_blocks
            
            logger.info(f"Функция: {func.__name__}")
            logger.info(f"Время выполнения: {execution_time:.4f} секунд")
            logger.info(f"Выделено блоков памяти: {blocks_used} блоков")
            
            if execution_time > TIME_THRESHOLD:
                logger.warning(f"ПРЕДУПРЕЖДЕНИЕ: {func.__name__} превысила порог времени "
                             f"({execution_time:.4f}с > {TIME_THRESHOLD}с)")
            
            if blocks_used > BLOCKS_THRESHOLD:
                logger.error(f"ОШИБКА: {func.__name__} превысила порог блоков памяти "
                           f"({blocks_used} блоков > {BLOCKS_THRESHOLD} блоков)")
                
            return result
            
        except Exception as e:
            logger.error(f"Ошибка в функции {func.__name__}: {str(e)}")
            raise
            
    return wrapper

# Тесты
@monitor_performance
def heavy_computation(n: int) -> list:
    """Тестовая функция с высокой вычислительной нагрузкой."""
    return [i**2 for i in range(n)]

@monitor_performance
def memory_intensive_operation(size: int) -> list:
    """Тестовая функция с большим потреблением памяти."""
    return ['x' * 100 for _ in range(size)]

logger.info("Запуск теста вычислительной функции...")
heavy_computation(1000000)
print()
    
logger.info("Запуск теста функции с потреблением памяти...")
memory_intensive_operation(50000)
print()
    
logger.info("Тестирование завершено")
```

```python
from functools import reduce, partial
from typing import Callable, Any

def compose(*functions: Callable) -> Callable:
    if not functions:
        return lambda x: x
    return reduce(
        lambda f, g: lambda *args, **kwargs: f(g(*args, **kwargs)),
        reversed(functions)
    )

def create_text_pipeline() -> Callable[[str], str]:
    def del_whitespace(text: str) -> str:
        return text.strip()

    def to_lower(text: str) -> str:
        return text.lower()

    def validate(text: str) -> str:
        if not text:
            raise ValueError("Строка не может быть пустой")
        for char in text:
            if not char.isalpha():
                raise ValueError(f"Недопустимый символ '{char}' — разрешены только буквы")
        return text
    
    return reduce(
        lambda f, g: lambda *args, **kwargs: g(f(*args, **kwargs)),
        [del_whitespace, to_lower, validate]
    )

def make_validator(allowed_chars: str) -> Callable[[str], str]:
    def validator(text: str) -> str:
        if not text:
            raise ValueError("Строка не может быть пустой")
        for char in text:
            if char not in allowed_chars:
                raise ValueError(f"Недопустимый символ '{char}' — разрешены только: {allowed_chars}")
        return text
    return validator

validate_only_latin = make_validator("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ")

def replace_invalid_with(text: str, invalid_char: str = '!') -> str:
    result = []
    for char in text:
        if char.isalpha():
            result.append(char)
        else:
            result.append(invalid_char)
    return ''.join(result)

validate_partial = partial(replace_invalid_with, invalid_char='!')

#тесты
compose()("test")

def add_one(x): return x + 1
def multiply_two(x): return x * 2
func = compose(add_one, multiply_two)
func(5)

pipeline = create_text_pipeline()
pipeline("  HelloWorld  ")

try:
    pipeline("Hello123")
except ValueError:
    pass

try:
    pipeline("   ")
except ValueError:
    pass

validator = make_validator("abc123")
validator("a1b2c3")

try:
    validator("abcd")
except ValueError:
    pass

validate_only_latin("HelloWorld")

try:
    validate_only_latin("Привет")
except ValueError:
    pass

replace_invalid_with("Hello123World!")
replace_invalid_with("Hello123", "_")

validate_partial("Test!123")

print("Все функции работают корректно")
```

Задание 7.1: Декораторы для измерения времени
Создайте декораторы для:

Кэширования результатов функции (мемоизация)
Ограничения частоты вызовов функции (rate limiting)

```python
import time
import functools
from collections import defaultdict

def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Функция выполнилась за {end_time - start_time:.4f} секунд")
        return result
    return wrapper

def count_calls(func):
    def wrapper(*args, **kwargs):
        wrapper.call_count += 1
        print(f"Функция вызвана {wrapper.call_count} раз")
        return func(*args, **kwargs)
    wrapper.call_count = 0
    return wrapper

def cache_results(func):
    cache = {}
    def wrapper(*args, **kwargs):
        key = str(args) + str(kwargs)
        if key not in cache:
            cache[key] = func(*args, **kwargs)
            print("Функция вычислена и сохранена в кэш")
        else:
            print("Результат функции взят из кэша")
        return cache[key]
    return wrapper

def rate_limit(max_calls, period):
    def decorator(func):
        calls = []
        def wrapper(*args, **kwargs):
            current_time = time.time()
            
            calls[:] = [call_time for call_time in calls if current_time - call_time < period]
            
            if len(calls) >= max_calls:
                wait_time = period - (current_time - calls[0])
                print(f"Превышен лимит вызовов. Подождите {wait_time:.2f} секунд")
                time.sleep(wait_time)

                current_time = time.time()
                calls[:] = [call_time for call_time in calls if current_time - call_time < period]
            
            calls.append(current_time)
            return func(*args, **kwargs)
        return wrapper
    return decorator

@measure_time
@count_calls
def slow_function(x):
    time.sleep(0.1)
    return x * 2

@cache_results
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

@rate_limit(max_calls=2, period=1)
def limited_function(x):
    print(f"Выполняется limited_function с аргументом {x}")
    return x + 1

for i in range(3):
    result = slow_function(i)
    print(f"Результат: {result}")
    
print("\n=== Тест cache_results (Фибоначчи) ===")
print(f"fibonacci(10) = {fibonacci(10)}")
print(f"fibonacci(10) = {fibonacci(10)}")  # Должно взять из кэша
print("\n=== Тест rate_limit ===")
for i in range(5):
    result = limited_function(i)
    print(f"Результат: {result}")
    time.sleep(0.3)
```
